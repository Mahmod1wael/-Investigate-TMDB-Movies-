## Data Cleaning


```python
data.drop(labels =['imdb_id', 'budget', 'revenue', 'cast', 'homepage', 'tagline', 'overview','vote_count', 'vote_average','runtime'], axis=1, inplace=True)
```

```python
data.shape
```

```python
data.drop_duplicates(inplace=True)
data.duplicated().sum()
```

```python
#changing rows with zeros
data['budget_adj'] = data['budget_adj'].replace(0, np.NaN)
data['revenue_adj'] = data['revenue_adj'].replace(0, np.NaN)
#dropping null values
data.dropna(inplace=True)
data.isna().sum()
```

```python
data['release_date'] = pd.to_datetime(data['release_date'])
data['release_date'].head()
```

# Exploratory Data Analysis



## Question 1: How are the budget of making a movie changed over time ?

```python
x=data.groupby('release_year').budget_adj.mean().index
y=data.groupby('release_year').budget_adj.mean().values
#size of plot
plt.figure(figsize=(18,9))
plt.bar(x,y);
#x axis label
plt.xlabel('Release Year',size=20)
plt.xticks(size=15)
#y axis label
plt.ylabel('Budget (Adj)',size=20)
plt.yticks(size=15)
#plot label
plt.title('Average Spending over Time',size=15);
sns.set_style("whitegrid")
```

## Question 2 : Which movies have the most and least profit?


```python
#making new column for profit
data['profit'] = data['revenue_adj'] - data['budget_adj']
#finding movies with highest and lowest profit 
most =  pd.DataFrame(data.loc[data['profit'].idxmax()])
least = pd.DataFrame(data.loc[data['profit'].idxmin()])
df_profit = pd.concat([most, least],axis=1)
df_profit
```

'Star Wars' movie Directed by George Lucas in 1977 has the most profit,

'The Warrior's Way' movie Directed by Sngmoo Lee in 2010 has the least profit (huge loss)

```python
data_mp = pd.DataFrame(data['profit'].sort_values(ascending = False))
data_mp['original_title'] = data['original_title'].copy()

#defining axes
movies = data_mp['original_title'][:10]
profit = data_mp['profit'][:10]

#size of plot  
fig = plt.figure(figsize = (18, 9))

# creating the barh plot
plt.barh(movies, profit, color ='black')

#x axis label
plt.xlabel("Profit",size=15)
plt.xticks(size=15)

#y axis label
plt.ylabel("Movies",size=15)
plt.yticks(size=15)

#plot label
plt.title("Top 15 movies with highest profit",size = 15)
sns.set_style("whitegrid")
```

from this plot we got most 15 movie with highest profit

 ## Question 5 : What is the number of movies released every year?


```python
#group data by 'release year' and 'id' count
data.groupby('release_year').count()['id'].plot(xticks = np.arange(1960,2016,5))
sns.set(rc={'figure.figsize':(18,9)})
#x label
plt.xlabel('Release year',fontsize = 15)
#y label
plt.ylabel('Number of Movies',fontsize = 15)
#plot label
plt.title("number of movies per year",fontsize = 15)
sns.set_style("whitegrid")
```

number of movies is increasing every year, dropped after 2011
## Question 6 : What is the number of movies for each genre?

```python
def split_values(x):
    #link all the rows of the genres.
    link_rows = data[x].str.cat(sep = '|')
    #split all values
    da = pd.Series(link_rows.split('|'))
    #return each genre and it's count.
    return da.value_counts(ascending=False)
```

```python
#call the function we called
genres = split_values('genres')

#plot with kind 'bar'
genres.plot(kind= 'bar',figsize = (18,9),fontsize=14,colormap='Pastel2')
#plot label
plt.title("Number of movies for each genre",fontsize=14)
#x axis label
plt.xlabel('Number Of Movies',fontsize=14)
#y axis label
plt.ylabel("Genres",fontsize= 14)
sns.set_style("whitegrid")
```

we notice that Drama is the most genre while TV movie is the least.
## Question 7 : Which director has the most popular movies?

```python

#group rows by directior and popularity column
director = data.groupby('director').popularity.sum().sort_values(ascending=False)[:15]
#plot with kind 'barh'
director.plot(kind= 'barh',figsize = (18,9),fontsize=14,colormap='Pastel1')

#plot label
plt.title("best 15 director",fontsize=14)
#x axis label
plt.xlabel('Number of Movies',fontsize=14)
#y axis label
plt.ylabel("Director",fontsize= 14)
sns.set_style("whitegrid")
```

Christopher Nolan has the most number of movies
## Question 8 : Which production companies have most movies?

```python

#call the function we created beore to split rows and return there counts
production_companies = split_values('production_companies')

#select first 15 value
production_companies.iloc[:15].plot(kind='barh',figsize=(18,9),fontsize=14)
#plot label
plt.title("Number of movies for each company",fontsize=14)
#x axis label
plt.xlabel('Number Of Movies',fontsize=14)
#y axis label
plt.ylabel('Company Name',fontsize=14)
sns.set_style("whitegrid")
```

##### From the plot above we notice that (Universal Pictures, Warner Bros) have the most number of movies
